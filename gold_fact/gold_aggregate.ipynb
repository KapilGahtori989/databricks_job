```python
%sql
CREATE OR REPLACE TABLE learning_spark_sql.ecommerse_data.gold_daily_revenue
USING DELTA
COMMENT 'Daily revenue by region'
AS
SELECT
  region,
  SUM(order_amount) AS total_revenue
FROM learning_spark_sql.ecommerse_data.gold_fact_table
GROUP BY region;
```
