```python
%sql
CREATE OR REPLACE TABLE learning_spark_sql.ecommerse_data.gold_fact_table
USING DELTA
COMMENT 'Fact table for e-commerce sales'
AS
SELECT
  o.order_id,
  o.order_timestamp,
  c.customer_name,
  c.country,
  p.product_name,
  p.category,
  o.region,
  o.quantity,
  o.order_amount
FROM learning_spark_sql.ecommerse_data.silver_orders o
LEFT JOIN learning_spark_sql.ecommerse_data.silver_customers c
  ON o.customer_id = c.customer_id
LEFT JOIN learning_spark_sql.ecommerse_data.silver_products p
  ON o.product_id = p.product_id;
```
