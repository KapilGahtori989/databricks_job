```python
%sql
CREATE OR REPLACE TABLE learning_spark_sql.ecommerse_data.silver_orders
USING DELTA
COMMENT 'Validated order transactions'
AS
SELECT
  order_id,
  CAST(order_timestamp AS TIMESTAMP) AS order_timestamp,
  customer_id,
  product_id,
  quantity,
  order_amount,
  region
FROM learning_spark_sql.ecommerse_data.bronze_orders
WHERE order_amount > 0
  AND quantity > 0;
```
