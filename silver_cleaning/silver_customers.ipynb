```python
%sql
CREATE OR REPLACE TABLE learning_spark_sql.ecommerse_data.silver_customers
USING DELTA
COMMENT 'Cleaned customer dimension'
AS
SELECT
  customer_id,
  customer_name,
  email,
  country,
  CAST(signup_date AS DATE) AS _date
FROM learning_spark_sql.ecommerse_data.bronze_customers
WHERE customer_id IS NOT NULL;

```
