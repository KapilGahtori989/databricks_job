```python
%sql
CREATE OR REPLACE TABLE learning_spark_sql.ecommerse_data.silver_products
USING DELTA
COMMENT 'Cleaned product dimension'
AS
SELECT
  product_id,
  product_name,
  category,
  price
FROM learning_spark_sql.ecommerse_data.bronze_products
WHERE price > 0;

alter table learning_spark_sql.ecommerse_data.silver_products
set tblproperties (
  'delta.columnMapping.mode' = 'name'
);
```
